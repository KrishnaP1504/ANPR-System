```python
from ultralytics import YOLO
import os
```

```python
# 1. Define path to your dataset's data.yamlr
dataset_yaml_path = os.path.join(os.getcwd(), 'dataset', 'data.yaml')
```

```python
# 2. Load the YOLOv11 model
model = YOLO('yolo11n.pt')
```

```python
# 3. Train the model
results = model.train(
    data=dataset_yaml_path,
    epochs=50,
    imgsz=640,
    batch=16,
    project='models',
    name='anpr_yolov12',
    device=0  # Use 0 for GPU, 'cpu' for CPU
)
```

```python
# 4. Validate the model
metrics = model.val()
print(f"Map50: {metrics.box.map50}")
```

```python
# 5. Export the model (Optional, but good for deployment)
success = model.export(format='onnx')
```
